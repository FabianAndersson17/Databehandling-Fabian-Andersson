# Merging datasets

```python
# setup
import pandas as pd
import numpy as np

df1 = pd.DataFrame(np.arange(16).reshape(4,4), columns=list("ABCD"))
df2 = pd.DataFrame(np.zeros((3,4)), columns=list("ABCD"))
df1
```

```python
df2
```

## Concat
- concatinates along a particular axis
- set logic

```python
pd.concat([df1, df2]).reset_index(drop = True)
```

```python
# df1 shape (4,4) df2 shape (3,4)
pd.concat([df1, df2], axis=1, join="inner")
```

## Merge
- Diffrent join operations similar to relational databases such as SQL

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K0", "K2", "K3"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)


right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)

left, right
```

```python
pd.merge(left, right, on="key", indicator=True)
```

```python
left.merge(right, on="key", how="outer", indicator=True)
```
